```python
#  ETAPA 1 - Importar dados e limpar

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados do dataset de vendas online (formato Excel)
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx'
df = pd.read_excel(url)

# Padronizar nomes das colunas para facilitar o uso posterior
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]

# Limpeza dos dados:
# - Remover registros com valores ausentes em 'customerid' e 'description'
# - Remover transações canceladas (InvoiceNo começando com 'C')
# - Manter apenas quantidades e preços positivos
# - Remover valores extremos de quantidade e preço unitário
# - Filtrar para dados do Reino Unido
# - Converter a coluna de data para o formato datetime
# - Criar nova coluna de preço total (quantidade x preço unitário)
df.dropna(subset=['customerid', 'description'], inplace=True)
df = df[~df['invoiceno'].astype(str).str.startswith('C')]
df = df[(df['quantity'] > 0) & (df['unitprice'] > 0)]
df = df[df['quantity'] < 10000]
df = df[df['unitprice'] < 10000]
df = df[df['country'] == 'United Kingdom']
df['invoicedate'] = pd.to_datetime(df['invoicedate'])
df['totalprice'] = df['quantity'] * df['unitprice']
df.drop_duplicates(inplace=True)
```

```python
#  ETAPA 2 - Análise gráfica básica

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 5)

# Criar coluna de data para análise por dia
df['date'] = df['invoicedate'].dt.date

# 1. Faturamento diário
faturamento_diario = df.groupby('date')['totalprice'].sum().reset_index()

plt.figure()
sns.lineplot(x='date', y='totalprice', data=faturamento_diario)
plt.title(' Faturamento Diário')
plt.xlabel('Data')
plt.ylabel('Total de Vendas (£)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Produtos mais vendidos
# Agrupar descrições de produtos por soma da quantidade
# Ordenar os mais vendidos e selecionar os 10 primeiros
top_produtos = (
    df.groupby('description')['quantity']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure()
sns.barplot(data=top_produtos, x='quantity', y='description')
plt.title(' Top 10 Produtos Mais Vendidos')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Produto')
plt.tight_layout()
plt.show()

# 3. Clientes que mais gastaram
top_clientes = (
    df.groupby('customerid')['totalprice']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure()
sns.barplot(data=top_clientes, x='customerid', y='totalprice', palette='flare')
plt.title(' Top 10 Clientes que Mais Gastaram')
plt.xlabel('ID do Cliente')
plt.ylabel('Total Gasto (£)')
plt.tight_layout()
plt.show()

# 4. Distribuição de vendas por hora do dia
df['hour'] = df['invoicedate'].dt.hour

plt.figure()
sns.histplot(df['hour'], bins=24, kde=False)
plt.title(' Distribuição de Vendas por Hora do Dia')
plt.xlabel('Hora')
plt.ylabel('Número de Vendas')
plt.tight_layout()
plt.show()

# 5. Relação entre quantidade vendida e preço unitário
plt.figure()
sns.scatterplot(x='quantity', y='unitprice', data=df, alpha=0.4)
plt.title(' Relação: Quantidade x Preço Unitário')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Preço Unitário (£)')
plt.xlim(0, 100)
plt.ylim(0, 50)
plt.tight_layout()
plt.show()

# 6. Estatísticas rápidas das principais variáveis numéricas
print("\nResumo estatístico geral:")
print(df[['quantity', 'unitprice', 'totalprice']].describe().round(2))
```

```python
#  ETAPA 3 - Estatística Básica

# Estatísticas descritivas para as variáveis principais
print("\n Estatísticas descritivas (Quantidade, Preço Unitário e Total da Venda):")
print(df[['quantity', 'unitprice', 'totalprice']].describe().round(2))

# Produto mais vendido (moda da descrição)
produto_mais_vendido = df['description'].mode()[0]
print(f"\n Produto mais vendido (moda): {produto_mais_vendido}")



# Top 5 clientes com maior faturamento
faturamento_clientes = (
    df.groupby('customerid')['totalprice']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)
print("\n Top 5 clientes que mais compraram:")
print(faturamento_clientes)



# Visualização de outliers com boxplots
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['quantity'], color='lightblue')
plt.title('📦 Distribuição de Quantidades Vendidas')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['totalprice'], color='salmon')
plt.title('💲 Distribuição de Valor Total por Venda')

plt.tight_layout()
plt.show()



# Coeficiente de variação para medir dispersão relativa
cv_quantity = df['quantity'].std() / df['quantity'].mean()
cv_totalprice = df['totalprice'].std() / df['totalprice'].mean()



print(f"\n Coeficiente de variação da quantidade: {cv_quantity:.2f}")
print(f" Coeficiente de variação do total da venda: {cv_totalprice:.2f}")
```

```python
# ETAPA 4 - Modelagem de Machine Learning: Clusterização de Clientes

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA



# Agrupar os dados por cliente e calcular totais de quantidade, total de compras e número de compras distintas
clientes = df.groupby ('customerid').agg({
    'quantity': 'sum',
    'totalprice': 'sum',
    'invoiceno': 'nunique'
}).rename(columns={'invoiceno': 'num_compras'}).reset_index()
clientes = df.groupby('customerid').agg({
    'quantity': 'sum',
    'totalprice': 'sum',
    'invoiceno': 'nunique'
}).rename(columns={'invoiceno': 'num_compras'}).reset_index()



# Padronizar os dados para clusterização
scaler = StandardScaler()
clientes_scaled = scaler.fit_transform(clientes[['quantity', 'totalprice', 'num_compras']])



# Aplicar algoritmo KMeans com 4 clusters definidos previamente
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
clientes['cluster'] = kmeans.fit_predict(clientes_scaled)



# Redução de dimensionalidade com PCA para visualização dos clusters
pca = PCA(n_components=2)
clientes_pca = pca.fit_transform(clientes_scaled)
clientes['pca1'] = clientes_pca[:, 0]
clientes['pca2'] = clientes_pca[:, 1]


 # Visualizar os clusters formados com base nas duas principais componentes
plt.figure(figsize=(10, 6))
plt.figure(figsize=(10, 6))
sns.scatterplot(data=clientes, x='pca1', y='pca2', hue='cluster', palette='Set2', s=100)
plt.title(' Segmentação de Clientes via K-Means')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()

# Estatísticas médias por grupo formado
print("\n Estatísticas por Cluster:")
resumo_cluster = clientes.groupby('cluster')[['quantity', 'totalprice', 'num_compras']].mean().round(2)
print(resumo_cluster)
```
